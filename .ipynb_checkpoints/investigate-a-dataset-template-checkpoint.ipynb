## Exploratory Data Analysis

### Research Question 1 : How many "Yes" and "No" no-show reponses was received? And can patients' gender be used to predict the overall appointment show-up responses?

```python
#total no-show("No") and no-show("Yes") response in the dataset.
yes_no_count = df['no_show'].value_counts()
print(yes_no_count)
```

```python
#to show the distribution of the no-show('No')and no-show('Yes')
df['no_show'].hist(color='b',alpha=0.8)
plt.xlabel('Response')
plt.ylabel('Number of Patients')
plt.title("Distribution of the no-show('Yes')and no-show('No')")
plt.legend()
#plt.show()
plt.xticks([0,0.9],['showed','No-showers'])
```

The above distribution tells that generally,a higher number of patients didn't miss their appointment.

```python
#how many females showed-up and how many didn't
df_f = df.query('gender=="F"')
f = df_f['no_show'].value_counts()

#how many males showed-up and how many didn't
df_m = df.query('gender=="M"')
m = df_m['no_show'].value_counts()
print(f,m)
```

```python
plt.bar(['M','F'],[30962,57246], color='g')
plt.title('Appointment no_show("No") count for both genders')
plt.xlabel('Gender')
plt.ylabel('Show-up count')
```

From the barchart above,more females than males showed-up on their appointment date.

```python
plt.bar(['M','F'],[7725,14594], color='g')
plt.title('Appointment no_show("Yes") count for both genders')
plt.xlabel('Gender')
plt.ylabel('no_show count')
```

Also,the barchart above shows that more females than males didn't show-up on their appointment date.

***Below is a function I created in order to avoid repetitive coding, as I'll be finding proportional values twice - now and later.***

```python
# a function to calculate proportion when called upon.
def calc_proportion(x,y) :
    prop_x = (x*100)/(x+y)
    prop_x = "{:.2f}".format(prop_x)
    prop_y = (y*100)/(x+y)
    prop_y = "{:.2f}".format(prop_y)
    return{"no_show No-response" : prop_x+'%',"no_show Yes-response" : prop_y+'%'}
```

```python
#to call a function to get the proportion of 'yes/no' response in percentage for all the females.
calc_proportion(57246,14594)
```

From the result of the function call above, 79.69% of females in the dataset showed up for appointment, while 20.31% of females didn't.

```python
#to call a function to get the proportion of 'yes/no' responses in percentage for all the males.
calc_proportion(30962,7725)
```

From the result of the function call above, 80.03% of males in the dataset showed up for appointment, while 19.97% of males didn't.

### From my research above :
* The percentage of females that showed-up for appointment is higher than the females who didn't.
* The percentage of males that showed-up for appointment is also higher than the males who did not.
* Overall,more number of females showed-up for appointments than males.
* Likewise overall,more number of females did not show-up for appointments than males.

Therefore, there is not enough evidence to say that the patients' genders can be used to predict the outcome of showing up or not on the appointment days.
